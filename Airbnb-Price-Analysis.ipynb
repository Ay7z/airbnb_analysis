checking and removing the outliers

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
```

```python
q1
```

```python

iqr = q3 - q1
```

```python
low_limit = q1 - 1.5*iqr
upp_limit = q3 + 1.5*iqr
```

```python
low_limit
```

```python
df = df[(df['price'] >= low_limit) & (df['price'] <= upp_limit)]
```

visualisations

```python
#distrubuiton of price
sns.histplot(df['price'] , bins = 50)
```

Insight:
Prices appear to be relatively evenly distributed across the selected
price range after preprocessing. No single price range dominates
the dataset, suggesting a broad spread of listing prices.

```python
#roomtype vs price
sns.boxplot(data = df, x = 'room_type', y = 'price')
plt.show()
```

 Insight:
 Although room types differ, their median prices remain relatively close.
This suggests that factors beyond room type, such as neighbourhood and
property features, may have a greater impact on Airbnb prices.

```python
# top locations
df['neighbourhood_group'].value_counts().plot(kind = 'bar')
```

Insight:
Hosts are significantly more active in Manhattan and Brooklyn, indicating
stronger tourism demand and greater earning potential compared to other.

```python
#co-relations
plt.figure(figsize = (10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap = 'Blues')
plt.show()
```

Insight:
Most features have weak correlations, indicating that Airbnb pricing is
likely influenced by multiple factors rather than a single variable.
Price and service fee exhibit a near-perfect positive correlation (0.99),
while review activity metrics show a moderate positive relationship (0.60).

```python
# dsitribution of room_type
df['room_type'].value_counts().plot.pie(autopct = '%1.1f%%', figsize = (6,6))
plt.show()
```

 Insight:
More than half of all listings are entire homes or apartments, making them
the most popular accommodation type. Private rooms also make up a significant
portion of the market, while shared rooms and hotel rooms are relatively rare.
This suggests that travelers generally prefer privacy and independent stays.

```python
#average price according to the room_type
sns.barplot(data = df, x= 'room_type', y = 'price')
plt.show()
```

Insight:
Hotel rooms have the highest average price among all accommodation types,
while private rooms and entire homes/apartments have very similar average prices.
Shared rooms fall in between, suggesting that room type alone does not create
large differences in pricing within this dataset.

```python
#total rooms according to room type
sns.countplot(data = df, x ='room_type', hue = 'neighbourhood_group')
plt.show()
```

Insight:
Manhattan and Brooklyn dominate the Airbnb market across all room types.
Entire homes/apartments are especially popular in Manhattan, while Brooklyn
shows a more balanced mix of private rooms and entire homes. Queens has a
moderate presence, whereas Bronx and Staten Island contribute only a small
share of total listings.

```python
#What is the overall price distribution of Airbnb listings, and how do hotel room prices compare to the overall market?

print("=== FULL PRICE STATS ===")
print(df['price'].describe())

print("\n=== HOTEL ROOM ONLY ===")
print(df[df['room_type'] == 'hotel room']['price'].describe())

print("\n=== PRICE RANGE ===")
print(f"Min: {df['price'].min()}")
print(f"Max: {df['price'].max()}")
print(f"Count of $0 or negative prices: {(df['price'] <= 0).sum()}")
```

Insight:
The average Airbnb listing price is $624, with half of all listings priced
between $338 and $912. Hotel rooms have a slightly higher average price of
$666, suggesting they command a premium over the overall market.

Key Findings:
Hotel rooms are priced slightly higher than the average Airbnb listing,
but the difference is not substantial. The absence of invalid price values
and the close alignment of mean and median prices indicate a clean and
well-structured dataset suitable for further analysis.
